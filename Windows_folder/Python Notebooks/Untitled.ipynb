```python
import quantiacsToolbox
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

```python
#reading file
F_AD = pd.read_csv(r'C:\Users\Anshul Goel\Desktop\Quantiacs\tickerData\F_AD.txt')
```

```python
#getting data of features for lookback no. of days
lback = 25
security=F_AD
CLOSE = np.array(security.loc[:lback-1, [' CLOSE']])
VOL = np.array(security.loc[:lback-1, [' VOL']])
OI = np.array(security.loc[:lback-1, [' OI']])
HIGH = np.array(security.loc[:lback-1, [' HIGH']])
LOW = np.array(security.loc[:lback-1, [' LOW']])
OPEN = np.array(security.loc[:lback-1, [' OPEN']])
#plt.plot(CLOSE)
#plt.plot(OPEN)
```

```python
#getting features of each
poly = PolynomialFeatures(degree=5)
close_ft = poly.fit_transform(CLOSE)
vol_ft = poly.fit_transform(VOL)
oi_ft = poly.fit_transform(OI)
high_ft = poly.fit_transform(HIGH)
low_ft = poly.fit_transform(LOW)
open_ft = poly.fit_transform(OPEN)
```

```python
#regression: obtaining for each
reg = linear_model.LinearRegression()
X = poly.fit_transform(np.arange(lback).reshape(-1, 1))
y = VOL[:]
reg.fit(X,y)
plt.plot(y)
plt.plot(reg.predict(X))
```
